#### Average Macronutrient Intake and its Relationship on Average Workout Duration in Hours
To analyze the relationship of macronutrient intake (proteins, carbs, fats) on workout session duration (hours), I binned the macronutrient variables, where each bin corresponds to a low, moderate, and high intake of that specific macronutrient. I chose to do this as it can create more understandable categories and can be easily understood when comparing low, moderate, and high intake diets with different variables. To determine the thresholds of what determines things like a "high protein" or a "moderate fat" intake, I focused on categorizing observations based on the percentage of macronutrient intake from the total caloric intake (e.g low protein diets correspond to diets with <10% of total calories derived from protein mean the same thing for someone consuming 1,500 calories versus 3,000 calories a day). I chose to do this method because realistically, everyone's bodily needs, caloric intake, and macronutrient intake are different. Using percentages instead of hardcoding values as the thresholds can evade subjectivity in what a low, moderate, or high intake is defined as.

For each macronutrient, I created the thresholds based off of Acceptable Macronutrient Distribution Ranges (AMDRs) mentioned in Medicine Libre Texts. This is a calculated range of how much energy from carbohydrate, fat, and protein is recommended for a healthy diet. The AMDRs are as follows: Protein 10-35%, Carbohydrate 45-65%, and Fat 20-35% [5]. 

```python
# Finding the percentage of calories for each macronutrient out of the total calorie intake
health_data['Protein_pct'] = (health_data['Proteins'] * 4) / health_data['Calories'] * 100 # 1g of protein is 4 calories
health_data['Carb_pct'] = (health_data['Carbs'] * 4) / health_data['Calories'] * 100 # 1g of carb is 4 calories
health_data['Fat_pct'] = (health_data['Fats'] * 9) / health_data['Calories'] * 100 # 1g of fat is 9 calories

# Creating the thresholds of macronutrients based on percentile distribution for binning
protein_bins = [0, 10, 35, np.inf]
carb_bins = [0, 45, 65, np.inf]
fat_bins = [0, 20, 35, np.inf]

labels = ['Low', 'Moderate', 'High']

# Creating the bins 
health_data['Protein_Bin'] = pd.cut(health_data['Protein_pct'], bins = protein_bins, labels = labels, include_lowest = True)
health_data['Carb_Bin'] = pd.cut(health_data['Carb_pct'], bins = carb_bins, labels = labels, include_lowest = True)
health_data['Fat_Bin'] = pd.cut(health_data['Fat_pct'], bins = fat_bins, labels = labels, include_lowest = True)
```

```python
# Thresholds for each macronutrient
pd.DataFrame({
    "Macronutrient": ["Protein", "Carbs", "Fats"],
    "Low–Moderate Threshold": [protein_bins[1], carb_bins[1], fat_bins[1]],
    "Moderate–High Threshold": [protein_bins[2], carb_bins[2], fat_bins[2]]})
```

For each macronutrient, the AMDR range categorizes the "moderate" intake category for each macronutrient. Observations where an individual who consume macronutrients below the AMDR threshold have a "low-intake", while individuals who consume macronutrients above the AMDR threshold have a "high-intake". For example, someone who consumes 10% of their total caloric intake derived from protein have a "low-intake" of protein. Similarly, an individual who eats above 35% of their total caloric intake derived from fat have a "high-intake" of fat, and so on. This method of focusing on percentages rather than setting rigid thresholds ensures that everyone's bodily needs are accounted for when categorizing low, moderate, or high intake of a specific macronutrient, instead of directly inputting an individual's macronutrient intake in grams. For example, a "high protein" diet does not mean the same for everyone. Eating 80g of protein may be considered "high-protein" for one person, but "moderate-protein" for another because it may take up different percentages of their total caloric intake for the day. This method ensures is more realistic and applicable to the real world based off of a great variance of lifestyle and bodily needs.

After finding the thresholds, I created the bins based off of distribution of each macronutrient, labeling them as "low", "moderate", and "high" based off of the AMDRs provided above.

```python
pd.DataFrame({
    "Low Intake Count": [
        health_data['Protein_Bin'].value_counts().get("Low", 0),
        health_data['Carb_Bin'].value_counts().get("Low", 0),
        health_data['Fat_Bin'].value_counts().get("Low", 0)],
    "Moderate Intake Count": [
        health_data['Protein_Bin'].value_counts().get("Moderate", 0),
        health_data['Carb_Bin'].value_counts().get("Moderate", 0),
        health_data['Fat_Bin'].value_counts().get("Moderate", 0)],
    "High Intake Count": [
        health_data['Protein_Bin'].value_counts().get("High", 0),
        health_data['Carb_Bin'].value_counts().get("High", 0),
        health_data['Fat_Bin'].value_counts().get("High", 0)]}, 
    index = ["Protein", "Carbs", "Fats"]).reset_index().rename(columns = {"index": "Macronutrient"})
```

Each macronutrient bin has a different number of observations with them, with a majority of observations  under the "moderate intake" category for both protein and fats. The carb bin has significantly more observations in the "low intake" category compared to other macronutrients and observations are more evenly distributed across the bins than protein and fats.

#### Analyzing Relationship Between Workout Experience Level and BMI
The Experience_Level variable is given as floats, ranging from 1 to 3.05. In the Kaggle dataset description, it states that 1 corresponds to beginner, 2 corresponds to intermediate, while 3 corresponds to advanced. Since the values are given as floats, I have decided to bin the Experience_Level variable and have the beginner group range from 1 to 1.99, the intermediate group range from 2 to 2.99, and advanced group include values 3 and higher. This method will allow for greater readability and understanding by grouping observations under a category based on the description of the dataset, instead of leaving them as floats which may be difficult for further analysis. This method can also avoid subjectivity in what defines a beginner, intermediate, or advanced person in fitness, and these thresholds are also explicitly stated in the dataset description.

```python
# Creating bins
bins = [1.0, 2.0, 3.0, bmi_filtered_q2['Experience_Level'].max()]
labels = ['Beginner', 'Intermediate', 'Advanced']

exp_cat, exp_bins = pd.cut(bmi_filtered_q2['Experience_Level'], bins = bins, labels = labels, include_lowest = True, retbins = True)
bmi_filtered_q2['Experience_Bin'] = exp_cat
 
# Create summary table
experience_bin_table = pd.DataFrame({
    'Workout Experience Level': labels,
    'Experience Level Range': [
        f"{bins[0]} – {bins[1]}",
        f"{bins[1]} – {bins[2]}",
        f"{bins[2]} – {bins[3]:.2f}"],
    'Number of Observations': bmi_filtered_q2['Experience_Bin'].value_counts().loc[labels].values})

display(experience_bin_table)
```

Looking at the number of observations across the experience levels, the "Beginner" workout experience level appears to have significantly more observations than the "Intermediate" and "Advanced" workout experience levels. To further visualize the relationship between experience level and health metrics, it is important to keep the number of observations under each bin into consideration in cases where there are greater variability and differences in statistical information. 

Now, I am going to focus on investigating the relationship between experience level with BMI using the data in which BMI < 16 was filtered out.

```python
sns.boxplot(data = bmi_filtered_q2, x = 'Experience_Bin', y = 'BMI', hue = 'Gender', palette = 'pink')
plt.xlabel('Workout Experience Level')
plt.ylabel('BMI')
plt.title('BMI vs. Workout Experience Level')
plt.legend(bbox_to_anchor = (1.3,1), title = 'Gender')
plt.show()
```

```python
def outlier_count(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return ((x < lower) | (x > upper)).sum()


summary = bmi_filtered_q2.groupby(['Experience_Bin', 'Gender'])['BMI'].agg(
    count = 'count',
    mean = 'mean',
    median = 'median',
    q1 = lambda x: x.quantile(0.25),
    q3 = lambda x: x.quantile(0.75)).reset_index()

summary['IQR'] = summary['q3'] - summary['q1']

summary['Outliers'] = (bmi_filtered_q2.groupby(['Experience_Bin', 'Gender'])['BMI'].apply(outlier_count).values)
summary = summary.rename(columns = {'Experience_Bin': 'Experience Level'})

summary
```

Looking at the boxplot, it appears that the average BMI stays consistent throughout each experience level across both males and females. For example, the average BMI for a male and female stays consistent at about 25 regardless of workout experience level. Similarly, the median bmi for both males and females across all levels of workout experience stays approximately around 24. This insight may be unexpected because BMI, defined as a quick screening tool that uses weight and height to estimate an individual's body fat, may be expected to have a more distinguishable relationship with workout experience (e.g. beginners may have higher BMIs due to just beginning their fitness journeys). However, this visualization suggests that the average BMI stays consistent across all levels of workout experience. Another thing to consider is the differing sizes of the interquartile range for each workout experience level and gender combination. Considering that the interquartile range represents the middle 50% of the data, it appears that the beginner experience level has greater variability in terms of BMI compared to intermediate or advanced experience levels. This observation can also be attributed to the fact that the beginner experience level group has signfiicantly more observations, which may contribute to greater variation of BMI in the interquartile range.

But, these extremely higher values are still achievable, but should not be ignored in attempts to mimic real-world scenarios with naturally high variation among individuals.

#### The Relationship Between Resting Heart Rate and Workout Frequency

The Workout_Frequency (days/week) variable is given as a float rather than an integer. It is possible that the values are reported as an average of workout frequency over an extended period of time. To account for this, I binned the Workout_Frequency (days/week) variable into low, moderate, and high frequency groups. The thresholds for these labels are subjective, so I tried to evenly space the groups out so each category covers 2-3 days out of the week. I chose to categorize [0,2) days a week as low frequency, [2, 4) days a week as moderate frequency, and 4 or more days a week as high frequency. Similar to binning experience level, I chose to bin workout frequency in this case to allow for greater readability and understanding by categorizing the values instead of leaving them as floats. I also chose to create low, moderate, and high frequency bins instead of 7 total bins representing working out 1-7 days a week due to the fact that the values are floats. For an example, if a value is reported as 2.5, it would be difficult to decide whether to label this individual as working out 2 days a week or 3 days a week. Therefore, I created low, moderate, and high frequency bins to avoid ambiguity.

```python
bins = [0,2,4,7]
labels = ['Low Frequency', 'Moderate Frequency', 'High Frequency']

exp_cat, exp_bins = pd.cut(bmi_filtered_q2['Workout_Frequency (days/week)'], bins = bins, labels = labels, include_lowest = True, retbins = True)
bmi_filtered_q2['Frequency_Bin'] = exp_cat
 
frequency_bin_table = pd.DataFrame({
    'Workout Frequency (days/week)': labels,
    'Workout Frequency Range': [
        f"{bins[0]} – {bins[1]}",
        f"{bins[1]} – {bins[2]}",
        f"{bins[2]} – {bins[3]:.2f}"],
    'Number of Observations': bmi_filtered_q2['Frequency_Bin'].value_counts().loc[labels].values})

display(frequency_bin_table)
```

Looking at the number of observations for each workout frequency (days/week) group, the moderate frequency group has significantly more observations than the low or high frequency group. This observation is worth considering when analyzing the visualizations with frequency and resting heart rate, as it is possible to have great variation or vastly different statistical information. For groups with lower numbers of observations, their averages are more vulnerable to drastic change from extreme values or outliers. While data cleaning, I observed that neither the Workout Frequency (days/week) variable or Resting_BPM variable have outliers, so this should not be a large concern.
